```python
import pandas as pd

# Load the dataset
data = pd.read_csv("dataset_med.csv")
```

```python
# Display basic information about the dataset
data_info = data.info()
data_head = data.head()

data_info, data_head
```

```python
# Convert date columns to datetime format
data['diagnosis_date'] = pd.to_datetime(data['diagnosis_date'], errors='coerce')
data['end_treatment_date'] = pd.to_datetime(data['end_treatment_date'], errors='coerce')
```

```python
# Check for duplicates and remove them if any
data = data.drop_duplicates()

# Summarize missing values
missing_values_summary = data.isnull().sum()

# Display the first few rows after date conversion and check for missing values
data.head(), missing_values_summary
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the label encoder
label_encoder = LabelEncoder()

# Columns to encode
categorical_columns = ['gender', 'country', 'cancer_stage', 'family_history', 
                       'smoking_status', 'treatment_type']

# Apply encoding to the selected columns
for column in categorical_columns:
    data[column] = label_encoder.fit_transform(data[column])

# Display the first few rows after encoding
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for plots (Use set_theme instead of set)
sns.set_theme(style="whitegrid")
```

```python
# Plot survival rates by cancer stage
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='cancer_stage', hue='survived', palette='viridis')
plt.title('Survival Rates by Cancer Stage', fontsize=16)
plt.xlabel('Cancer Stage', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Survived', loc='upper right')
plt.show()

```

```python
# Plot survival rates by treatment type
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='treatment_type', hue='survived', palette='coolwarm')
plt.title('Survival Rates by Treatment Type', fontsize=16)
plt.xlabel('Treatment Type', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Survived', loc='upper right')
plt.show()
```

```python
# Distribution of age by survival status
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='age', hue='survived', kde=True, palette='Set2', bins=30)
plt.title('Age Distribution by Survival Status', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.show()
```
